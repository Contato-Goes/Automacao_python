```python
from selenium import webdriver
navegador = webdriver.Chrome()
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from time import sleep

servico = Service(ChromeDriverManager().install())


# Url do site e do gerador de boletos
url = "https://tiburcio-estacionamento.herokuapp.com/"
url_gerador = "https://tiburcio-estacionamento.herokuapp.com/boleto/listar"

# Informações de login
usuario = "aldirgoes@gmail.com"
senha = "al274002"


# Entrando no navegador 
navegador.get(url)
navegador.maximize_window()
janela_principal = navegador.window_handles[0]

# Preencher campos com dados
navegador.find_element('xpath',
                      '/html/body/div/div[2]/div[2]/div/div/div[2]/form/div/div[1]/input').send_keys(usuario)
navegador.find_element('xpath',
                      '/html/body/div/div[2]/div[2]/div/div/div[2]/form/div/div[2]/input').send_keys(senha)
navegador.find_element('xpath',
                       '/html/body/div/div[2]/div[2]/div/div/div[2]/form/div/div[3]/div[2]/button').click()
sleep(1)

# Entrando no gerador de boletos
navegador.get(url_gerador)

sleep(1)

select = Select(navegador.find_element(By.ID, 'opcaoPesquisa'))
select.select_by_value('data')

sleep(1)

dia = input("Qual sera o dia ? ")
select = Select(navegador.find_element(By.ID, 'dataVencimento'))
select.select_by_value(dia)

sleep(1)
navegador.find_element('xpath',
                        '//*[@id="collapseOpcaoBoleto"]/div/div/div[3]/button').click()

options = Options()
options = webdriver.ChromeOptions()
options.add_experimental_option('prefs', {
#Change default directory for downloads
"download.default_directory": "D:\Imagens\Boletos\Boletos dia 8{}".format(dia), 
"download.prompt_for_download": False, #To auto download the file
"download.directory_upgrade": True,
"plugins.always_open_pdf_externally": True #It will not show PDF directly in chrome
})

sleep(3)
                                              
lista_boleto = navegador.find_element('xpath', '//*[@id="acordion-lista-boleto"]')
tamanho_lista = int(lista_boleto.get_attribute("childElementCount")) + 1

sleep(2)       
N1 = 1
while N1 < int(tamanho_lista):  
    sleep(2)  
    print(N1, tamanho_lista)
    cliente = navegador.find_element('xpath',
                        '//*[@id="acordion-lista-boleto"]/div[{}]'.format(N1)) 
    cliente_numero = cliente.get_attribute('wire:key')
    print(cliente_numero)  
    
    navegador.switch_to.new_window('tab')
    navegador.get('https://tiburcio-estacionamento.herokuapp.com/downloadFile/{}'.format(cliente_numero))                    
    sleep(3)
    N1 += 1
    navegador.close()
    
    navegador.switch_to.window(janela_principal)
    
    #tratamento de erro:
    navegador.get(url_gerador)
    #Selecionando data
    select = Select(navegador.find_element(By.ID, 'opcaoPesquisa'))
    select.select_by_value('data')
    sleep(3)
    select = Select(navegador.find_element(By.ID, 'dataVencimento'))
    select.select_by_value(dia)
    navegador.find_element('xpath',
                        '//*[@id="collapseOpcaoBoleto"]/div/div/div[3]/button').click()
    sleep(3)
```

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from time import sleep

servico = Service(ChromeDriverManager().install())

# Url do site e do gerador de boletos
url = "https://web.whatsapp.com/"
url_gerador = "http://estacionamento-tiburcio.herokuapp.com/gerar-boleto"
options = webdriver.ChromeOptions()
navegador = webdriver.Chrome(service=servico, options=options)

# Entrando no navegador 
navegador.get(url)

sleep(25)

# Preencher campos com dados
navegador.find_element('xpath',
                       '//*[@id="main"]/footer/div[1]/div/span[2]/div/div[1]/div[2]/div/span/div/div/ul/li[4]/button/span').click()
```
